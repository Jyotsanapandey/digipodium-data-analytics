```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [21,343,56,89,90,33,58,94]
plt.plot(y)
plt.show()
```

```python
plt.figure(figsize = (3,4))
plt.pie(y)
plt.show()      # show() function will hide the unnecessary story
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2 , figsize = (15,2))
plt.show()
```

```python
t = np.arange(-3 * np.pi, 3 * np.pi , 0.2)
x = np.sin(t)
y = np.cos(t)
fig, ax = plt.subplots(nrows = 1 , ncols = 2 , figsize = (15,2))
ax[0].plot(t,x,'r')
ax[1].bar(t,y)
plt.show()
```

```python
t = np.arange(-3 * np.pi, 3 * np.pi , 0.2)
x = np.sin(t)
y = np.cos(t)
fig, ax = plt.subplots(nrows = 1, ncols = 2 , figsize = (15,2))
ax[0].plot(t,x,'r')
ax[1].scatter(t,y)
plt.show()
```
