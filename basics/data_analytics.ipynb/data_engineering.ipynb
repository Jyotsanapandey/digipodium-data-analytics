### Dropping the column

```python
df.drop(columns=['Unnamed: 0'], inplace = True)
```

```python
df.head()
```

```python
df.replace('alfa-romero','alfa-romeo',inplace = True)
```

```python
df
```

```python
df.replace('?',np.nan , inplace = True)    # removes all the ? by .nan
```

```python
df
```

```python
df.dropna(subset=['price'] , inplace= True)
```

```python
df.info()
```

```python
df.isnull().sum()   # count the missing data in each column
```

```python
df['normalized-losses'].astype('float').sum()
```

```python
df['normalized-losses'].astype('float').mean()    # here we have calculated mean
```

```python
mnl = df['normalized-losses'].astype('float').mean() 
```

```python
df['normalized-losses'].replace(np.nan, mnl, inplace = True)
df.head()
```

```python
df
```

```python
df['bore'].astype('float').mean() 
```

```python
abc = df['bore'].astype('float').mean() 
```

```python
df['bore'].replace(np.nan, abc, inplace = True)
df.head()
```

```python
df['num-of-doors'].value_counts()           # shows the repeated value
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
```
